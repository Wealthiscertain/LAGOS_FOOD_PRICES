**4. OUTLINE THE REQUIRED ROWS:** In this analysis, we will only be making use of the 4 different species of Rice.

```python
lagos_rice_prices = pd.read_csv(filename,
                                usecols = ['id', 'ItemLabels', '20-Dec','21-Nov', '21-Dec'],
                               index_col =  'id',
                                skipfooter = 20,
                                engine ='python'
                               )

lagos_rice_prices = lagos_rice_prices [19:23]

lagos_rice_prices
```
